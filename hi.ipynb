```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import zipfile
from pprint import pprint # to get a more easily-readable view.
import re
pd.set_option('display.max_colwidth', None)
```

```python
review = pd.read_csv("data.csv", usecols= [3,5,6,7,9])
review.head()
```

```python
senti = pd.read_table("hw/hw2/vader_lexicon.txt",names = ['index','polarity'], index_col = 0, usecols = [0,1])
senti.head()
```

```python
review["reviewText"] = review["reviewText"].str.lower()
review.head()
```

```python
punct_re = r'[\W\s]'
def punctuation(text):
    return re.sub(punct_re, " ", str(text))


review['no_punc'] = review['reviewText'].apply(punctuation)
review.head()
```

```python
tidy_format = pd.DataFrame(review['no_punc'].str.split(expand = True))
tidy_format = tidy_format.stack()
tidy_format = tidy_format.reset_index(level = 1)
tidy_format.index.name = ''
tidy_format.columns = ['num','word']
tidy_format.head()
```

```python
review['polarity'] = tidy_format.join(senti, on='word').groupby(by=tidy_format.index.name).sum()['polarity']
review['polarity'] = review['polarity']/len(review['no_punc'])
review.head()
```
